```python
# import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# read cleaned data into a df
train = pd.read_csv('C:\\Users\\ashwani\\Project_ML_Finance/train_data2.csv')
test = pd.read_csv('C:\\Users\\ashwani\\Project_ML_Finance/test_data2.csv')

train.head()
```

```python
train.shape
```

```python
cols = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed',
       'Credit_History', 'Property_Area']

train = pd.get_dummies(train, columns = cols)
test = pd.get_dummies(test, columns = cols)
train = train.sample(frac=1).reset_index(drop=True)
train.head()
```

```python
test = test.sample(frac=1).reset_index(drop=True)
test.head()
```

```python
y = train.Loan_Status
```

```python
train =  train.drop(['Loan_ID', 'Loan_Status'], axis=1)
X = train.astype(np.float32)
Loan_Id = test['Loan_ID']
```

```python
test = test.drop(['Loan_ID'], axis=1)
test.insert(8, 'Married_Unknown', 0)
```

```python
#Splitting the given train dataset into train and Validation datasets for training and validating the model
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8)
```

```python
X_train
```

```python
X_val
```

```python
#Applying Logistic Regression Model on the validation dataset (splitted from given train dataset) for prediction of Loan Status 
#to measure the accuracy (Actual Vs. Predicted)
logreg = LogisticRegression(C=1e9) 
logreg.fit(X_train, y_train)

#print(logreg.coef_)
#print(logreg.intercept_)

y_pred = logreg.predict(X_val)
```

```python
from sklearn import metrics
print (metrics.accuracy_score(y_val,y_pred))
```

```python
#Comparison of Actual (Val Dataset) Vs. Predicted (Val Dataset)
actual = pd.DataFrame (y_val.values[0:])
predicted = pd.DataFrame (y_pred[0:])
Compare = pd.concat([actual, predicted], names=['Actual', 'Predicted'], axis=1)
Compare.columns = ['Actual', 'Predicted']
Compare
```

```python
#Applying Logistic Regression Model on given test dataset for prediction of Loan Status
logreg = LogisticRegression(C=1e9) 
logreg.fit(X, y)

#print(logreg.coef_)
#print(logreg.intercept_)

preds = logreg.predict(test)
#print (preds)
```

```python
#Depiction of the Loan Status Prediction for test dataset
preds = pd.Series(preds)
final_preds = pd.concat([Loan_Id, preds], names=['Loan_ID', 'Loan_Status'], axis=1)
final_preds.columns = ['Loan_ID', 'Loan_Status']
final_preds
```
